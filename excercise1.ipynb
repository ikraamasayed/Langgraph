```python
from typing import Dict, TypedDict
from langgraph.graph import StateGraph
```

```python
class AgentState(TypedDict):
    message : str
```

```python
def appreciation(state:AgentState):
    state['message'] = f"{state['message']}, You're  doing amazing Job"
    return state
```

```python
graph = StateGraph(AgentState)
graph.add_node("appreciate",appreciation)
graph.set_entry_point("appreciate")
graph.set_finish_point("appreciate")

app = graph.compile()
```

```python
from IPython.display import display,Image
display(Image(app.get_graph().draw_mermaid_png()))
```

```python
result = app.invoke({"message":"Ikrama"})
result['message']
```
